```python
# import dependencies
import pandas as pd
```

```python
#import measurements csv
measure_file = 'Resources/hawaii_measurements.csv'

measure_df = pd.read_csv(measure_file)
measure_df.head()
```

```python
#Looking at data with NaNs and missing values
measure_df.describe()
```

```python
measure_group = measure_df.groupby(["station"])
measure_avg = measure_group.mean()

measure_avg
```

```python
#drop missing values to see how it affects data
drop_measure_df = measure_df.dropna(how='any')
drop_measure_df.count()
```

```python
drop_measure_df.describe()
```

```python
drop_measure_group = drop_measure_df.groupby(["station"])
drop_measure_avg = drop_measure_group.mean()

drop_measure_avg
```

I chose to drop the NaN rows in my clean data set. The dataset is so large that we still have a good sampling without this information. While we lose temperature data by doing this, the trade off is that we do not then manipulate the precipation data by adding zeros. Since we do not know what the precipitation amount should be, adding zeros is adding false measurements that throw off our data. Removing the NaNs does change our data as well, but the data in the data set are then all still valid and the sample size is still large enough.

```python
#write to clean csv
drop_measure_df.to_csv("Resources/clean_hawaii_measurements.csv", index=False, header=True)
```
